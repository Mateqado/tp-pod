Creese una cuenta en Kaggle e importe los archivos del dataset del siguiente link: https://www.kaggle.com/datasets/usdot/flight-delays. Cree los dataframes `airlines`, `airports`, y `flights` apartir de esos archivos.

Combine (*join*) las tablas `airlines`, `airports`, y `flights` en una sola tabla.

```python
import pandas as pd

airlines = pd.read_csv('airlines.csv')
airports = pd.read_csv('airports.csv')
flights = pd.read_csv('flights.csv')

flights_airlines = pd.merge(flights, airlines, left_on='AIRLINE', right_on='IATA_CODE', how='left')
all_data = pd.merge(flights_airlines, airports, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', how='left')
all_data = pd.merge(all_data, airports, left_on='DESTINATION_AIRPORT', right_on='IATA_CODE', how='left', suffixes=('_origin', '_dest'))
```

¿Cuantos vuelos fueron al aeropuerto JFK?



```python
vuelos_JFK = all_data[all_data['DESTINATION_AIRPORT'] == 'JFK'].shape[0]

print(f"Numero de vuelos que fueron a JFK: {vuelos_JFK}")
```

¿Cuantos vuelos hizo la aerolinea AA?

```python
vuelos_AA = all_data[all_data['AIRLINE_x'] == 'AA'].shape[0]

print(f"Numeros de vuelos que hizo AA: {vuelos_AA}")
```

¿Que aerolineas (las primeras 5) tuvo la menor cantidad de vuelos con atrasos? Imprimirlas.

```python
vuelos_retrasados = all_data[all_data['ARRIVAL_DELAY'] > 0]

aerolineas_retrasadas = vuelos_retrasados.groupby('AIRLINE_x')['ARRIVAL_DELAY'].count()

aerolineas = aerolineas_retrasadas.sort_values()

top_5_fewest_delays = aerolineas.head(5)

print("Las 5 aerolineas con menos catidad de vuelos con atrasos son:")
print(top_5_fewest_delays)
```

¿Que aerolineas (las primeras 5) tuvo la mayor cantidad de vuelos con atrasos? Imprimirlas.

```python
aerolineas = aerolineas_retrasadas.sort_values(ascending=False)

top_5 = aerolineas.head(5)

print("Las 5 aerolineas con mayor cantidad de vuelos con atrasos")
print(top_5)
```

Haga un resumen de las razones por la cual los vuelos se atrasan.

```python
rasones = flights[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']].sum()

print("Motivos de retraso:")
print(rasones)

vuelos_tot = rasones.sum()
porcentajes = (rasones / vuelos_tot) * 100
print("\nMotivos de retraso en porcentaje:")
print(porcentajes)
```

```python
columnas = ['DEPARTURE_DELAY', 'ARRIVAL_DELAY', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY', 'LATITUDE_origin', 'LONGITUDE_origin', 'LATITUDE_dest', 'LONGITUDE_dest']

for col in columnas:
    all_data[col] = all_data[col].fillna(all_data[col].mean())

categorias = ['TAIL_NUMBER', 'CANCELLATION_REASON', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY', 'AIRPORT_origin', 'CITY_origin', 'STATE_origin', 'COUNTRY_origin', 'AIRPORT_dest', 'CITY_dest', 'STATE_dest', 'COUNTRY_dest']

for col in categorias:
    all_data[col] = all_data[col].fillna(all_data[col].mode()[0])
```
